```python

class KMeans:
    pass 
class Centroids:
    pass 

class RBFN:
    def __init__(self, Xtrain, Ytrain, Xtest, Ytest, hidden_neuron = 10) -> None:
        self.weights = np.zeros(hidden_neuron)
        self.hidden_neuron = hidden_neuron
        choices = np.random.choice(np.arange(Xtrain.shape[0]))
        self.centers = np.array(list(map(lambda x:Xtrain[x], choices)))
        
    def get_std(self, X_train):
        return np.std(X_train)
    
    def get_euclidean_distance(self, X1, X2):
        return np.linalg.norm(X1 - X2)
    
    def get_guassian_distance(self, X, mu):
        dis = self.get_euclidean_distance(X, mu)
        return np.exp(-(dis)**2)
    
    def get_weighted_sum(self, X):
        std = self.get_std()
        weighted_sum = 0
        for hidden_N in range(self.hidden_neuron):
            guassian = self.get_guassian_distance(X, self.centers[hidden_N])
            weight = self.weights[hidden_N]
            weighted_sum += guassian*weight 
        return weighted_sum/(2*std**2)
    
    def train(self, Xtrain, Ytrain, Xtest, Ytest, epochs = 100, lr = 0.01):
        Yhat = np.array([])
        n = Xtrain.shape[0]
        bias = 0
        for iter in range(epochs):
            #We can a threshold
            for X in Xtrain:
                output_score = self.get_weighted_sum(X)
                np.append(Yhat, output_score)
            E = Ytrain - Yhat
            dw = (2/n)*np.sum(X*E)
            db = (2/n)*np.sum(E)
            self.weights = self.weights + lr*dw 
            bias = bias + lr*db 
            print(f"weights: {iter + 1} iteration ===> ", self.weights, f"Training error: {iter + 1} ====> ", E)
            
            
            
# e(n) = d(n) - z(n)
# w(n +1) = w(n) +he(n)y(n)
    
        
    
        
            
       
    
    
        
    
        
```
